# **Essentials**

```python
#importing libraries
import pandas as pd
import numpy as np

#loading the data
df = pd.read_csv('Lending-company.csv')
```

```python
#checking the data
df.head()
```

#**1. Drop 'LoanID' and 'StringID' columns**

```python
#using drop function to drop the columns
df = df.drop(['LoanID', 'StringID'], axis=1)
```

```python
#checking if the columns are dropped
df.head()
```

# **2. Map Product categories**

```python
#mapping product categories or in other words changing them from categorical to numerical manually
df['Product'] = df['Product'].map({'Product A': 0, 'Product B': 1, 'Product C': 2, 'Product D': 2, 'Product E': 2, 'Product F': 2})
```

```python
#checking if the values are changed
df.head()
```

# **3. Replace 'NotSpecified' in Gender with mode**

```python
#find the mode of the Gender column
gender_mode = df['CustomerGender'].mode().iloc[0]
```

```python
#replace 'NotSpecified' with the mode
df['CustomerGender'] = df['CustomerGender'].replace('NotSpecified', gender_mode)
```

```python
#verify the changes
print(df['CustomerGender'].value_counts())
print(f"Mode of Gender column: {gender_mode}")
```

# **4. Extract Location and Region numbers**

```python
#changing the values from categorical to float
df['Location'] = df['Location'].str.extract('(\d+)').astype(float)
df['Region'] = df['Region'].str.extract('(\d+)').astype(float)
df.head()
```

# **5. Fill missing values in 'TotalPrice' with mean**

```python
#filling missing values with mean
df['TotalPrice'] = df['TotalPrice'].fillna(df['TotalPrice'].mean())
```

```python
#verifying changes
df.head()
```

# **6. Handle date-related operations**

```python
#changing startdate column to month and weekday columns
df['StartDate'] = pd.to_datetime(df['StartDate'], errors='coerce')
df = df.dropna(subset=['StartDate'])
df['Month'] = df['StartDate'].dt.month
df['Weekday'] = df['StartDate'].dt.dayofweek
df = df.drop('StartDate', axis=1)
```

```python
#verifying changes
df.head()
```

```python
#checking if eveything is fine with totaldaysyr column
print(df['TotalDaysYr'].describe())
```

# **8. Handle loan status**

```python
#handling loan status
df = df.dropna(subset=['LoanStatus'])
df = df[df['LoanStatus'] != 'Blocked']
df['LoanStatus'] = df['LoanStatus'].map({'Active': 0, 'Finished Payment': 1})

#display the first few rows and info of the processed dataframe
print(df.head())
print(df.info())
```
